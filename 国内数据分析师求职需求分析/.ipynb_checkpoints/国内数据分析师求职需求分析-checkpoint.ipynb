## 去掉实习职位

```python
x=df['position'].str.contains('实习')
df=df[~x]
df.reset_index(drop=True,inplace=True)
```

```python
df['address'].unique()
```

## 查看salary字段

```python
df['salary'].unique()
```

### 将薪资字段分为2段，bottom和top

```python
df['bottom'] = df['salary'].str.extract('(\d+).*')
df['top'] = df['salary'].str.extract('.*?-(\d+).*')
df.info()
```

```python
df
```

### 有些公司标注1X薪等，故新增一列commission_pct计算每个岗位的奖金率

```python
df['commission_pct'] = df['salary'].str.extract('.*?·(\d{2})薪')
df['commission_pct'].fillna(12,inplace=True)
df['commission_pct'] = df['commission_pct'].astype('float64')
df['commission_pct'] = df['commission_pct']/12
```

### 把bottom，top，commission__pct列转换为数值，计算出每个岗位的平均薪资avg_salary

```python
df['top'] = df['top'].astype('int64')
df['bottom'] = df['bottom'].astype(int)
df.info()
```

```python
df['avg_salary'] = (df.bottom+df.top)/2*df.commission_pct
df.head()
```

```python
df.describe()
```

```python
result = df['avg_salary'].value_counts()
result = pd.DataFrame(result)
result.reset_index(inplace=True)
result.columns = ['avg_salary','size']
result
```

## 对requires进行处理，分为学历和经验两列

```python
df.requires.unique()
```

```python
# 正则
df['diploma'] = df['requires'].str[-2:]
df['experience'] = df['requires'].str.extract('^(\d+?-?\d+年)')
df.head()
```

```python
# 对经验列数据进行修改
df.experience.replace(np.nan,'经验不限',inplace=True)
df.experience.replace('10年','10年以上',inplace=True)
df.experience.unique()
```

```python
# 对学历列数据进行修改
df.diploma.replace('不限','学历不限',inplace=True)
df.diploma.replace('中技','中专/中技',inplace=True)
df.diploma.unique()
```

```python
# 删除冗余列
df.drop(labels={'requires','salary'}, axis=1, inplace=True)
```

```python
d = df['avg_salary'].groupby(df['diploma'])
d_avg = d.mean()
c = list(d_avg.index)
v = list(range(1,len(c)+1))
w = d_avg.values.astype('int64')
x = d.median().values.astype('int64')
move = 0.4
plt.figure(figsize=(14,8),dpi=80)
plt.bar(v,w,width=move,color='#eed777')
plt.bar([i+move for i in v],x,width=move,color='#334f65')
a = np.arange(0,7)+1.2
plt.xticks(a,c)
plt.yticks(list(range(0,40,5)))
plt.legend(['均值','中位数'])
plt.title('各学历薪资均值及中位数比较图',fontsize=16)
plt.xlabel('学历',fontsize = 12)
plt.ylabel('薪资(单位K)',fontsize = 12)
for e,f in zip(v,w):
    plt.text(e,f,'{}k'.format(f),ha='center',fontsize=12)
for g,h in zip([i+move for i in v],x):
    plt.text(g,h,'{}k'.format(h),ha='center',fontsize=12)
plt.show()
```

从学历看是否是本科，对薪资水平的影响很大。
本科与硕士之间并没有特别大的差距，博士的薪资水平远远高于其他学历。

```python
data = df['diploma'].value_counts()
y=data.values
plt.figure(figsize=(10,10),dpi=80)
plt.pie(y,labels=data.index,autopct='%.1f %%')
plt.show()
```

从市场需求来看，要求本科占据绝大多数，要求硕士的较少，而博士则是极少数
